Continua amb el conjunt de dades de tema esportiu i genera una mostra utilitzant el mètode Reservoir sampling.

+ *Utilitzo l'algorisme que hi ha al recurs "The 5 Sampling Algorithms every Data Scientist need to know"*

```python
import random
def generator(max):
    number = 1
    while number < max:
        number += 1
        yield number

stream = tenis_df['match_minutes']

k=100
reservoir = []
for i, element in enumerate(tenis_df['match_minutes']):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
           
             reservoir[random.choice(range(0,k))] = element
```

```python
reservoir
```

+ *I comprovo que tinc els 100 valors esperats quan hem fixat k = 100 i l'estadística bàsica  de la variable*

```python
len(reservoir)
```

```python
a, b, c = min(reservoir), max(reservoir), sum(reservoir)/len(reservoir)
print('El valor mínim es {} el valor màxim és {} i la duració mitjana dels partits és de {}'.format(a, b, c))
```
